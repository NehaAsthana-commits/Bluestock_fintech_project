```python
df_transactions = pd.read_csv('../data/raw/08_investor_transactions.csv')

print("Shape:", df_transactions.shape)
print("\nData types:\n", df_transactions.dtypes)
print("\nFirst 5 rows:")
df_transactions.head()
```

```python
print("Missing values per column:\n", df_transactions.isna().sum())

print("\n--- AMFI Code Validation ---")
valid_codes = set(df_fund_master['amfi_code'])
transaction_codes = set(df_transactions['amfi_code'])
invalid_codes = transaction_codes - valid_codes

print("Number of unique amfi_codes in transactions:", len(transaction_codes))
print("Number of invalid amfi_codes (not in fund_master):", len(invalid_codes))
print("Sample invalid codes:", list(invalid_codes)[:10])

invalid_rows = df_transactions[df_transactions['amfi_code'].isin(invalid_codes)]
print("Number of transaction rows with invalid amfi_code:", len(invalid_rows))
print("% of total transactions:", round(len(invalid_rows) / len(df_transactions) * 100, 2), "%")
```

```python
df_holdings = pd.read_csv('../data/raw/09_portfolio_holdings.csv')

print("Shape:", df_holdings.shape)
print("\nData types:\n", df_holdings.dtypes)
print("\nFirst 5 rows:")
df_holdings.head()
```

```python
print("Missing values per column:\n", df_holdings.isna().sum())

# Check if weight_pct sums to ~100% per fund
weight_check = df_holdings.groupby('amfi_code')['weight_pct'].sum()
print("\nWeight % sum per fund (sample):")
print(weight_check.head(10))
print("\nFunds where total weight is far from 100%:")
print(weight_check[(weight_check < 90) | (weight_check > 110)])
```

```python
valid_codes = set(df_fund_master['amfi_code'])
holdings_codes = set(df_holdings['amfi_code'])
invalid_holdings_codes = holdings_codes - valid_codes

print("Unique amfi_codes in holdings:", len(holdings_codes))
print("Invalid amfi_codes (not in fund_master):", len(invalid_holdings_codes))
print("Sample invalid codes:", sorted(invalid_holdings_codes)[:10])

invalid_holding_rows = df_holdings[df_holdings['amfi_code'].isin(invalid_holdings_codes)]
print("Number of holding rows with invalid amfi_code:", len(invalid_holding_rows))
print("% of total holdings rows:", round(len(invalid_holding_rows) / len(df_holdings) * 100, 2), "%")
```
